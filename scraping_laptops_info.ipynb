```python
hrefs = []
for i in range(1, 10):
    url = ""
    if i == 1:
        url = f"https://www.technodom.kz/catalog/noutbuki-i-komp-jutery/noutbuki-i-aksessuary/noutbuki"
    else:
        url = f"https://www.technodom.kz/catalog/noutbuki-i-komp-jutery/noutbuki-i-aksessuary/noutbuki?page={i}"
    response = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
    html = urlopen(response)
    bs = BeautifulSoup(html, 'html.parser')
    ul = bs.find_all('ul', class_="category-page-list__list")
    li = ul[0].find_all('li')
    i = 0
    for item in li:
        if item.find('a') == None: continue
        hrefs.append(item.find('a').get('href'))
print(len(hrefs))
```

```python
for i in range(0, 20):
    print(hrefs[i] + " " + str(i))
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
import time

videocards = []
memory_sizes = []
memory_types = []
rams = []
laptop_series = []
classes = []
resolutions = []
matrix_types = []
weights = []

cpu_performances = []
manufacturers = []
cpu_cores = []
cpu_frequency = []
cpu_series = []

prices = []

options = webdriver.ChromeOptions()
  
driver = webdriver.Chrome(executable_path="chromedriver",
                              options=options)

def fetchInformation(start, end):
    for i in range(start, end):
        # Send a get request to the url
        laptop_url = f'https://www.technodom.kz/{hrefs[i]}'

        
       
        if (i == 5 or i == 8) and start == 0:
            continue
            
        driver.get(laptop_url)
        if i == 0:
            time.sleep(10)
            
            driver.find_element('xpath','/html/body/div[5]/div/div/div/button[1]').click()        
            
        driver.find_element(by=By.TAG_NAME, value='body').send_keys(Keys.PAGE_DOWN)

        time.sleep(3)
        try:
            driver.find_element('xpath','/html/body/div[1]/section/main/section/article/div/div[1]/div[1]/div[2]/div[2]/div[2]/div[2]/button').click()

        except NoSuchElementException:
            driver.find_element('xpath', '/html/body/div[1]/section/main/section/article/div/div[1]/div[1]/div[2]/div[2]/div[3]/div[2]/button').click()

        except:
            continue
            
        
            
        bs = BeautifulSoup(driver.page_source, 'html.parser')
        
        desired_divs = bs.find_all('div', class_='technical-specifications__attribute product-description')
        
        price_div = bs.find('div', class_='product-info__prices product-prices')
        price_p = price_div.find('p')
        prices.append(price_p.text.replace('\xa0', '')[:-1])
        
        p_texts = []
        for div in desired_divs:
            if div.find('p').text == 'Основные характеристики':
                p_texts = div.find_all('p', class_='Typography product-description__right-text Typography__Body Typography__Body_Small')
                if len(p_texts) == 1:
                     laptop_series.append("")
                     classes.append(p_texts[0].text)
                else:
                    laptop_series.append(p_texts[0].text)
                    classes.append(p_texts[1].text)
                    print('Основные характристики', i)
                    
            
            if div.find('p').text == 'Память':
                p_texts = div.find_all('p', class_='Typography product-description__right-text Typography__Body Typography__Body_Small')
                memory_sizes.append(p_texts[0].text)
                memory_types.append(p_texts[1].text)
                rams.append(p_texts[2].text)
                print('Память', i)

            if div.find('p').text == 'Видео':
                 p_texts = div.find_all('p', class_='Typography product-description__right-text Typography__Body Typography__Body_Small')
                 videocards.append(p_texts[0].text)
                 print('Видео', i)
            if div.find('p').text == 'Дисплей':
                 p_texts = div.find_all('p', class_='Typography product-description__right-text Typography__Body Typography__Body_Small')
                 resolutions.append(p_texts[0].text)
                 matrix_types.append(p_texts[-1].text)
                 print('Дисплей', i)
            if div.find('p').text == 'Габариты':
                 p_texts = div.find_all('p', class_='Typography product-description__right-text Typography__Body Typography__Body_Small')
                 weights.append(p_texts[-2].text)
                 print('Габариты', i)
            if div.find('p').text == 'Процессор':
                 p_texts = div.find_all('p', class_='Typography product-description__right-text Typography__Body Typography__Body_Small')
                 manufacturers.append(p_texts[0].text)
                 
                 if 'поколение' in p_texts[-1].text or 'Ryzen' in p_texts[-1].text:
                     cpu_frequency.append(p_texts[-2].text)
                     cpu_series.append(p_texts[-3].text)
                     cpu_cores.append(p_texts[2].text)
                 elif p_texts[0].text == 'Apple':
                     cpu_frequency.append("")
                     cpu_cores.append(p_texts[1].text)
                     if 'M' in p_texts[-1].text:
                         cpu_series.append(p_texts[-1].text)
                     else:
                         cpu_series.append(p_texts[-2].text)
                 else:
                     cpu_frequency.append(p_texts[-1].text)
                     cpu_cores.append(p_texts[2].text)
                     cpu_series.append(p_texts[-2].text)
                 print('Процессор', i)
            

fetchInformation(0, len(hrefs))

# Teacher, do not pay attention to these prints, they were added in order to find where some 
# kind of error occurs. Although the code is the same, but while scraping a lot of pages, 
# the Internet connection and other factors that do not depend only on the code begin to play a role.
```

```python
df = pd.DataFrame(laptops_dict)
df.to_csv('laptops_info.csv')
```
